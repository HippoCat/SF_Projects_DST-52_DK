# 11. Какого жанра фильмов больше всего?

```python
# Option 1

#--------1---------2---------3---------4---------5---------6---------7---------8
# Create empty dictionary to store unique genres
genres_dict = {}

# Number of movies in our dataset
movies_number = len(data)

# Fill our dictionary with unique genres using setdefault method
i = 0
while i < movies_number:
    for genre in data.iloc[i]['genres']:
        genres_dict.setdefault(genre, 0)
    i += 1

# Now we count the number of each genre in our dataset
i = 0
while i < movies_number:
    for genre in data.iloc[i]['genres']:
        genres_dict[genre] += 1 
    i += 1

# Create dataframe object from our dictionary by index
unique_genres_df = pd.DataFrame.from_dict(genres_dict,
                                          orient='index',
                                          columns=['count'])

# Create new dataframe sorted by genres count
unique_genres_df.sort_values('count', ascending=False)
```
